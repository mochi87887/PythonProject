# 比較運算子
用來比較「兩個值大小」的運算子，運算的結果只有 True 或是 False 兩種，常用於邏輯判斷使用

```python
a = 5
b = 3

print(a < b)
print(b <= a)
print(a != b)
print(a == b)
```

# 邏輯運算子
」可以判斷*「and」( 且 )、「or」( 或 ) 和「not」( 非 ) *三種邏輯狀態，運算的結果只有 True 或是 False 兩種

and

*   a 是 True，b 也是 True，結果是 True
*    a、b 其中一個是 False，結果就是 False



```python
a = True
b = False
c = True

print(a & b)
print(a and b)
print(a & c)
print(a and c)
```

or

只要 a、b 其中一個是 True，結果就是 True

```python
a = True
b = False
c = True

print(a | b)
print(a or b)
print(a | c)
print(a or c)
```

not

*   a 為 True，not a 的結果 False
*   a 為 False，not a 的結果 True



```python
a = True
b = False

print(not a)
print(not b)
```

# in 與 is 運算子

*   使用「in 運算子」可以判斷 b 是否包含 a
*   使用「is 運算子」可以判斷 a 和 b 是否為相同物件

判斷包含或相同，回傳 True，否則回傳 False



```python
a = 2
b = 4
c = [1,2,3]
print(a in c)
print(b in c)

x = [1,2,3]
y = [1,2,3]
z = x
print(x is y)
print(x is z)
```

# 位元運算子
針對每個數值的「二進位」字元進行位元運算

*    4 ( 二進位 0100 )
*    5 ( 二進位 0101 )


詳細參考文章用法


```python
print(4 & 5)
print(4 | 5)
print(4 ^ 5)
print(~4)
print(4 >> 2)
print(5 << 2)
```

# 跨列運算子

「算式過長，需要換行」的程式碼或運算式
*   使用小括號包覆使其換行計算
*   使用「反斜線 \」放在一列的最後方



```python
a = ( 1 + 2 + 3 +
      4 + 5 + 6 +
      7 + 8 + 9 )

b = 1 + 2 + 3 + \
    4 + 5 + 6 + \
    7 + 8 + 9

print (a)
print (b)
```
