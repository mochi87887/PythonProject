# int(x)
可以將數字或者是字串的參數，轉換為「整數」的型別


1.   數字有小數點，會無條件捨去小數點的數字
2.   x 為空，則返回 0
3.   True 和 False 則會轉換為 1 和 0



```python
a = int('9487')
b = int(87.878787)
c = int()
d = int(True)
e = int(False)
print(a)
print(b)
print(c)
print(d)
print(e)
```

# float(x)
可以將整數或者是字串的參數，轉換為「浮點數」的型別 ( 帶有小數點的數字 )


1.   數字為整數，會自動在後方加上「.0」
2.    x 為空，則返回 0.0
3.   True 和 False 則會轉換為 1.0 和 0.0



```python
a = float('123')
b = float(456)
c = float()
d = int(True)
e = int(False)
print(a)
print(b)
print(c)
print(d)
print(e)
```

# abs(x)
可以回傳 x 的「絕對值」

```python
a = abs(-789)
print(a)
```

# divmod(x, y)
回傳一個內容為「( x//y, x%y ) 的 tuple」( x 除以 y 的商和餘數 )

```python
a = divmod(5,3)
b = divmod(9,2)
print(a)
print(b)
```

# max(iter)
回傳一個可迭代物件 ( 字串、tuple、串列 ) 中的「最大值」

```python
a = max('100200300')
b = max([400,500,600])
c = max((700,800,900))
print(a)
print(b)
print(c)
```

# min(iter)
回傳一個可迭代物件 ( 字串、tuple、串列 ) 中的「最小值」

```python
a = min('100200300')
b = min([400,500,600])
c = min((700,800,900))
print(a)
print(b)
print(c)
```

# pow(x, y, z)


1.   使用 pow(x, y) 會回傳「x 的 y 次方」
2.   使用 pow(x, y, z) 會回傳「x 的 y 次方除以 z 的餘數」



```python
a = pow(2, 3)
b = pow(2, 3, 3)
print(a)
print(b)
```

# round(x, y)
1.   回傳「四捨五入後的 x，y 表示四捨五入的小數點位數」
2.   不指定 y 則表示四捨五入後的結果為整數



```python
a = round(3.14159)
b = round(3.14159, 3)
print(a)
print(b)
```

# sum(iter, y)

1.   回傳「串列或 tuple 的數值與 y 的加總」
2.   不指定 y 則 y 為 0



```python
a = sum([1,2,3,4])
b = sum((1,2,3,4))
c = sum((1,2,3,4), 5)
print(a)
print(b)
print(c)
```

# bool(x)
將參數 x 轉變成布林值 False 或 True


1.   內容是 0 或空值就會是 False
2.   其他內容就會是 True




```python
a = bool(1)
b = bool(0)
c = bool()
d = bool(999)
e = bool('hello')
f = bool([0])
g = bool([])
print(a)
print(b)
print(c)
print(d)
print(e)
print(f)
print(g)
```
