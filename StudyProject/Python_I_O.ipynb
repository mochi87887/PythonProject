# print(*objects, sep, end, file, flush)
print() 是 Python 負責將結果「輸出」的函式

1.   objects 要印出的對象，可以用逗號分隔多個對象
2.   sep 預設為「一個空白」，表示每個分隔的對象，在印出時中間的間隔符號
3.   end 預設換行符「\n*」，表示印出的結尾符號
4.   file 預設「None」，表示要寫入的文件對象
5.   flush 預設「False」，設定為 True 可以防止函式對輸出資料進行緩衝，並強行重新整理



```python
print(10,20,30)
print(11,21,31,sep=';')
print(12,22,32,sep=';',end='!')   #有設定 end，則會取代換行符，導致下一個 print 的結果和上一個連在一起
print(13,23,33)
```

# Loading 讀取效果

```python
# 「Loading」後方的「.」每隔 0.5 秒出現，若將 flush 移除或改為 False，會發現結果一次出現六個「.」
# 可重複執行觀看
import time
print('Loading',end = '')
for i in range(6):
    print('.',end = '',flush = True)
    time.sleep(0.5)
```

# 倒數計時效果

```python
# print 的內容前方加上「\r」的命令，就能讓每次印出時的游標位置，移動到該行的開頭
# 搭配 end 不換行的方式，就能做出類似「畫面更新」的效果
# 可重複執行觀看
import time
n = 10
for i in range(n+1):
    print(f'\r倒數 {n-i} 秒', end='')
    time.sleep(1)
print('\r時間到', end='')
```

# input(x)
執行 input(x) 後，命令列會顯示 x 內容，並等待使用者輸入內容，按下 enter 之後再進行下方的動作

```python
text = input('輸入文字：')
print('你輸入的是：'+ text)
```

# 輸入兩個數字，顯示兩個數字加總的結果

```python
num1 = int(input('輸入第一個數字：'))
num2 = int(input('輸入第二個數字：'))
print(f'兩個數字相加結果是：{num1 + num2}')
```
