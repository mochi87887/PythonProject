# if 判斷
如果判斷為 True，就會執行對應的程式，
如果判斷為 False，就會跳過判斷式，繼續執行下方的程式

```python
a = 2
b = 3

if a > b:
    print('Yes')
print('No')
```

# if、else

1.   if 對應結果為 True 的程式，else 對應結果為 False 的程式
2.   if 對應結果為 False 的程式，else 對應結果為 True 的程式



```python
a = 2
b = 3

if a > b:
    print('Yes')
else:
    print('No')
print('Ok')
```

# if、elif、else

1.   一個邏輯判斷裡，只會有一個 if 和 else，但可以有多個 elif
2.   不論是 if、elif 還是 else，最後只會有一種結果



```python
a = 2
b = 3

if a > b:
    print('a > b')

elif a < b:
    print('a < b')

else:
    print('a = b')
print('Ok')
```

```python
# 遇到「不想執行任何動作」的狀況，可以使用「pass」作為空式子，藉以保持語法的正確性
a = 2
b = 3

if a > b:
    pass

elif a < b:
    print('a < b')

else:
    print('a = b')
print('Ok')
```

# 巢狀判斷
「一個判斷式裡，還有另外 n 個判斷」，就像鳥巢一般層層判斷下去

```python
a = 2
b = 3

if a < b:
    print('a < b')

    if a == 1:
        print('a = 1')

    elif a == 2:
        print('a = 2')

    elif a==3:
        print('a = 3')

elif a > b:
    print('a > b')

else:
    print('a = b')
print('Ok')
```

# 三元運算式 ( 條件運算式 )

1.   一個判斷式裡面只有 if 和 else，就能夠使用三元運算式來簡化
2.   三元運算式的語法為：變數 = 值1 if 條件式 else 值2









```python
a = 1
b = 2
c = ''

if a > b:
    c = 'a'
else:
    c = 'b'
print(c)
```

```python
a = 1
b = 2
c = ''
c = 'a' if a > b else 'b'
print(c)
```
