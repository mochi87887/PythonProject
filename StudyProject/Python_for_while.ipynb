# for 迴圈
「for 迴圈」的使用的方法為「for 變數 in 可迭代的物件:」

```python
for i in 'abc':
    print(i)    # a  b  c ( 字串 )

for i in ['a','b','c']:
    print(i)    # a  b  c ( 串列 )

for i in {'a','b','c'}:
    print(i)    # c  b  a ( 集合 )

for i in {'a':1,'b':2,'c':3}:
    print(i)    # a  b  c ( 字典 )
```

# 巢狀迴圈
1.   將 for 迴圈放在 for 迴圈裡
2.   巢狀迴圈會先從「最內層」的迴圈開始執行，執行完畢後再執行外層的迴圈



```python
for a in ['x','y','z']:
    for b in [1,2,3]:
        print(b)
print(a)
```

```python
for a in ['x','y','z']:
    for b in [1,2,3]:
        print(f'{a}{b}')
```

# while 迴圈
1.   「根據條件判斷，決定是否重複或停止」的迴圈，用法為「while 條件:」
2.   如果條件判斷為 True，就會不斷執行迴圈內容，如果判斷為 False，就會停止迴圈



```python
a = 1
while a <= 5:
    print(a)
    a += 1
```

# break 和 continue
操作迴圈的過程中，可以使用「break 中斷」和「continue 跳出」兩個方法來「停止迴圈」

1.    break 會將整個迴圈停止
2.    continue 是將迴圈目前執行的程式停止，然後再次執行迴圈



break

```python
# 內層迴圈使用了 break，碰到 b 為 1 的時候就會停止，就看不見 2 和 3
for a in ['x','y','z']:
    for b in [1,2,3]:
        if(b == 2):
            break
        print(f'{a}{b}')
print('Ok')
```

continue

```python
for a in ['x','y','z']:
    for b in [1,2,3]:
        if(b == 2):
            continue
        print(f'{a}{b}')
print('ok')
```
